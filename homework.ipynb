```python
spotify = spotify[['track_name', 'artist(s)_name', 
        #'artist_count', 
        'released_year',
        'released_month', 
        #'released_day', 
        #'in_spotify_playlists',
        'in_spotify_charts', 'streams', 
        #'in_apple_playlists', 'in_apple_charts',
        #'in_deezer_playlists', 'in_deezer_charts', 'in_shazam_charts', 'bpm',
        #'key', 'mode', 
        'danceability_%', #'valence_%', 
        'energy_%',
        #'acousticness_%', 'instrumentalness_%', 'liveness_%', 'speechiness_%
        ]].copy()

#drops multiple columns from the list
```

```python
spotify['streams'] = pd.to_numeric(spotify['streams'], errors='coerce', downcast='integer')
#changes streams from an object into a float type
```

```python
spotify.rename(columns={'track_name':'Track_Name',
                        'artist(s)_name':'Artist(s)_Name',
                        'released_year':'Released_Year',
                        'released_year':'Released_Year',
                        'in_spotify_charts':'Spotify_Charts',
                        'streams':'Streams',
                        'danceability_%':'Danceability_%',
                        'energy_%':'Energy_%'})
#renames columns
```

```python
spotify.isna().sum()
#tells how many NaN (missing values) there are in each column
```

```python
spotify.loc[spotify.duplicated()]
#tells how many duplicated data there is and where
```

```python
ax = spotify['released_year'].value_counts() \
    .head(10) \
    .plot(kind='bar', title='Top Years Songs Introduced')
ax.set_xlabel('Year Released')
ax.set_ylabel('Number of Songs')

#creates a barchart to portray data
```

```python
spotify.plot(kind='scatter',
             x='released_year',
             y='streams',
             title='Relationship between Year Released and Streams')
```

```python
sns.scatterplot(x='released_year',
                y='streams',
                data=spotify,
                hue='energy_%')

#using seaborn
```
